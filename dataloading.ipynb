```python
import pandas as pd
import matplotlib.pyplot as plt
headers = ['Date', 'Open', 'High', 'Low', 'Close', 'Adj Close', 'Volume']
filePath = 'datasets/NVDAHistoric.csv'
```

```python
df = pd.read_csv(filePath, names = headers)
df_info = df.info()
df_head = df.head()

df_info, df_head
```

```python
# Reload the CSV file using semicolon as the delimiter
df_fixed = pd.read_csv(filePath, delimiter=';', names=headers)

# Display the first few rows and data types of the columns again
df_fixed_info = df_fixed.info()
df_fixed_head = df_fixed.head()

df_fixed_info, df_fixed_head
```

```python
# Remove dollar signs and commas, and convert columns to numeric types
columns_to_clean = ['Open', 'High', 'Low', 'Close']

# Remove non-numeric characters and convert to float
df_fixed[columns_to_clean] = df_fixed[columns_to_clean].replace({'\$': '', ',': ''}, regex=True).apply(pd.to_numeric, errors='coerce')


# Display the cleaned data
df_cleaned_head = df_fixed.head(100)


df_cleaned_head
```

```python
x = df_fixed['Date']    # X-axis data (e.g., Date)
y = df_fixed['Close']
y_open = df_fixed['Open']
df.set_index('Low').plot()
plt.figure(figsize=(10, 6))  # Optional: set figure size
plt.plot(x, y, label='Close Price')  
plt.plot(x, y_open, label='Open Price')
plt.xlabel('Date')  # Set x-axis label
plt.ylabel('Close Price')  # Set y-axis label
plt.title('Date vs Close Price')  # Set the title of the plot
plt.legend()  # Show legend
plt.grid(True)  # Show grid
plt.show()  # Display the plot
```
