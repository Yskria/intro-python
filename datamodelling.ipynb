```python
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as sStats
import seaborn as sn
pd.set_option('display.max_rows', None)
filePath = 'datasets/StudentsPerformance.csv'

studentPerforamance  = pd.read_csv(filePath)

readingScore = studentPerforamance['reading_score']
mathScore = studentPerforamance['math_score']
writingScore = studentPerforamance['writing_score']

plt.subplot(2, 1, 1)
plt.scatter(readingScore, writingScore)
plt.title('Reading Score vs Writing Score')
plt.xlabel('Reading Score')
plt.ylabel('Writing Score')
sn.regplot(x=readingScore, y=writingScore, scatter_kws = {'color':'green'}, line_kws={'color':'pink'})

plt.subplot(2, 1, 2)
plt.scatter(readingScore, mathScore)
plt.title('Reading Score vs Math Score')
plt.xlabel('Reading Score')
plt.ylabel('Math Score')

plt.tight_layout()
sStats.pearsonr(readingScore, mathScore)
```
